```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# Load Dataset
df = pd.read_csv("student_data.csv")
df.head()

```

```python
# Finding number of rows and columns
print("Shape (rows, columns): ", df.shape, "\n")

print("Number of rows: ", df.shape[0])
print("Number of columns: ", df.shape[1])
```

```python
# Check for nan values
print(df.isna().sum())
```

```python
# Check if there is duplicate values
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Viewing the data types of columns
df.dtypes
```

```python
# Display statistical summary
df.describe(include='all')
```

```python
# Plot the distribution of final grades (G3)
plt.figure(figsize=(8,5))
sns.histplot(df['G3'], bins=20)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Count")
plt.show()
```

Most students scored between 10 and 15, indicating average academic performance with few extreme low or high scores.

```python
# Plot the distribution of study time categories
plt.figure(figsize=(8,5))
sns.histplot(df['studytime'], bins=4)
plt.title("Distribution of Study Time")
plt.xlabel("Study Time Category")
plt.ylabel("Count")
plt.show()
```

Most students fall into study time category 2, indicating moderate study habits among the majority.

```python
# Plot the distribution of student absences
plt.figure(figsize=(8,5))
sns.histplot(df['absences'], bins=20)
plt.title("Distribution of Absences")
plt.xlabel("Number of Absences")
plt.ylabel("Count")
plt.show()
```

Most students have few absences, while a small number show very high absenteeism, resulting in a right-skewed distribution.

```python
# Plot the distribution of past academic failures
plt.figure(figsize=(8,5))
sns.histplot(df['failures'], bins=4)
plt.title("Distribution of Past Failures")
plt.xlabel("Number of Failures")
plt.ylabel("Count")
plt.show()
```

The majority of students have no past failures, while only a small proportion have one or more previous failures.

```python
# Visualize the relationship between past failures and final grade
plt.figure(figsize=(8,5))
sns.boxplot(x='failures', y='G3', data=df)
plt.title("Past Failures vs Final Grade")
plt.xlabel("Number of Past Failures")
plt.ylabel("Final Grade (G3)")
plt.show()
```

Students with more past failures tend to have lower final grades, indicating a negative relationship between previous failures and academic performance.

```python
# Visualize the relationship between study time and final grade
plt.figure(figsize=(8,5))
sns.boxplot(x='studytime', y='G3', data=df)
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.show()
```

Students who study more tend to achieve higher final grades, showing a positive relationship between study time and academic performance.

```python
# Visualize the relationship between absences and final grade
plt.figure(figsize=(8,5))
sns.scatterplot(x='absences', y='G3', data=df)
plt.title("Absences vs Final Grade")
plt.xlabel("Number of Absences")
plt.ylabel("Final Grade (G3)")
plt.show()
```

There is a slight negative relationship between absences and final grade, as students with higher absences generally tend to score lower.
